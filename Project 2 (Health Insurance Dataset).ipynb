# Encoding

```python
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()
```

```python
df_health['region']=lb.fit_transform(df_health['region'])
```

```python
df_health['sex']=lb.fit_transform(df_health['sex'])
df_health['smoker']=lb.fit_transform(df_health['smoker'])
```

# Removing outliers

```python
from scipy.stats import zscore
zs=np.abs(zscore(df_health))
threshold=3
print(np.where(zs>3))
D_F=df_health[(zs<3).all(axis=1)]
D_F
```

```python
plt.boxplot(D_F['bmi'])
plt.show()
```

```python
X=D_F.drop('charges',axis=1)
X
```

```python
Y=D_F.iloc[:,-1]
Y
```

# Spliting data set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Building Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
predi=lr.predict(X_test)
```

```python
print(predi,Y_test)
```

```python
from sklearn import metrics
metrics.r2_score(Y_test,predi)
```
