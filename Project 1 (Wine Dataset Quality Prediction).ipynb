```python
from imblearn.over_sampling import SMOTE
smp=SMOTE()
x, y = smp.fit_resample(x,y)
```

```python
y.value_counts()
```

# Artibitrary cut off

```python
def binary (p):
    if p>=7:
        return 1
    else:
        return 0
   
```

```python
y=y.apply(binary)
```

```python
y
```

# scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
LR=LogisticRegression()
SV=SVC()
DTC= DecisionTreeClassifier()
RC= RandomForestClassifier()
KNN= KNeighborsClassifier()

model=[LR,SV,DTC,RC,KNN]



for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    pred=m.predict(x_test)
    
    print('Accuracy score :', m)
    print(accuracy_score(y_test,pred))   
    print(confusion_matrix(y_test,pred)) 
    print(classification_report(y_test,pred))      
```

# Hyperparameter Tuning

```python
param={'criterion':['gini','entropy']}
gd=GridSearchCV(DTC,param)

gd.fit(x_train,y_train)

```

```python
gd.best_params_
```

```python
Best_Model = DecisionTreeClassifier(criterion= 'entropy')
Classifier = Best_Model.fit(x_train, y_train)
BM_pred = Best_Model.predict(x_test)
BM_acc = accuracy_score(y_test, BM_pred)
print("Accuracy score for the Best Model is:", BM_acc)
```
